# Importing python libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("hotel_bookings.csv")
```

# Removing Duplicate Values

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
              
```

```python
df.duplicated().sum()
```

```python
df.shape
```

# Handling Missing Values

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
plt.show()
```

```python
null_columns = ['agent', 'children', 'company']
for col in null_columns:
    df[col] = df[col].fillna(0)

# Replacing NA values with 'others'
    df['country'] = df['country'].fillna('others')
```

```python
df.isnull().sum()
```

# # Exploratory Data Analysis

# Distribution of hotel types

```python
sns.countplot(data=df, x='hotel')
plt.title('Distribution of Hotel Types')
plt.show()
```

# Distribution of booking channels

```python
sns.countplot(data=df, x='distribution_channel')
plt.title('Distribution of Booking Channels')
plt.show()
```

# Distribution of canceled bookings

```python
sns.countplot(data=df, x='is_canceled')
plt.title('Distribution of Canceled Bookings')
plt.show()
```

# Distribution of customer types

```python
sns.countplot(data=df, x='customer_type')
plt.title('Distribution of Customer Types')
plt.show()
```

# Length of stay distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='stays_in_weekend_nights', bins=20, kde=True, color='blue', label='Weekend Nights')
sns.histplot(data=df, x='stays_in_week_nights', bins=20, kde=True, color='red', label='Week Nights')
plt.title('Length of Stay Distribution')
plt.xlabel('Length of Stay')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Average daily rate by hotel type

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='hotel', y='adr', estimator=sum)
plt.title('Average Daily Rate by Hotel Type')
plt.ylabel('Average Daily Rate')
plt.xlabel('Hotel Type')
plt.show()
```

# Boxplot of lead time

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='hotel', y='lead_time')
plt.title('Lead Time Distribution by Hotel Type')
plt.ylabel('Lead Time')
plt.xlabel('Hotel Type')
plt.show()
```
